```python
#LeetCode 225. Implement Stack using Queues

class listQueue:
    def __init__(self):
        self.__queue=[]

    def enqueue(self, x):
        self.__queue.append(x)

    def dequeue(self):
        self.__queue.pop(0)

    def front(self):
        return self.__queue[0]

    def isEmpty(self) -> bool:
        return not bool(self.__queue)

    def dequeueAll(self):
        self.__queue.clear()

class MyStack:

    def __init__(self):
        self.q1=listQueue()
        self.q2=listQueue()
        

    def push(self, x):
        self.q2.enqueue(x)
        while not self.q1.isEmpty():
            self.q2.enqueue(self.q1.dequeue())
        self.q2, self.q1=self.q1, self.q2
        

    def pop(self):
        return self.q1.dequeue()
        

    def top(self):
        return self.q1.front()
        

    def empty(self) -> bool :
        return not bool(self.q1)
        
stack=MyStack()
stack.push(1)
stack.push(2)
stack.top()
stack.pop()
stack.empty()
```

```python
#Leetcode 232

class listStack:
    def __init__(self):
        self.__stack=[]

    def push(self, x):
        self.__stack.append(x)

    def pop(self):
        self.__stack.pop()

    def isEmpty(self) :
        return not bool(self.__stack)

    def top(self):
        if self.isEmpty() :
            return None
        else:
            return self.__stack[-1]
            
    def bottom(self):
        if self.isEmpty():
            return None
        else:
            return self.__stack[0]

class MyQueue:

    def __init__(self):
        self.s1=listStack()
        self.s2=listStack()
        

    def enqueue(self, x):
        self.s1.push(x)
        

    def dequeue(self):
        while not self.s1.isEmpty() :
            self.s2.push(self.s1.pop())
            
        self.s2.pop()
        
        while not self.s2.isEmpty() :
            self.s1.push(self.s2.pop())

    def front(self):
        if self.s1.isEmpty():
            return None
        else:
            return self.s1.bottom()
        

    def isEmpty(self):
        return not bool(self.s1)
        

queue=MyQueue()
queue.enqueue(1)
queue.enqueue(2)
queue.front()
queue.dequeue()
queue.isEmpty()
```
